```python
import cv2

# Load original image
original = cv2.imread("parte32.jpg")

# Image to classify point of contact
marked = cv2.imread("parte8.jpg")

#cropping to the bottom left block Vehicle A
original = original[115:280, 70:320] 
marked = marked[115:280, 70:320] 


# Find difference between the two images
diff = cv2.absdiff(original, marked)

# Convert difference to grayscale
diff_gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)

# Threshold the difference image to isolate the accident location
ret, thresh = cv2.threshold(diff_gray, 40, 255, cv2.THRESH_BINARY)

# Find coordinates of the mark
coords = cv2.findNonZero(thresh)

# Calculate center of the mark
moments = cv2.moments(coords)
center_x = int(moments["m10"] / moments["m00"])
center_y = int(moments["m01"] / moments["m00"])

# Draw center of the mark on the marked image (red dot)
cv2.circle(marked, (center_x, center_y), 5, (0, 0, 255), -1)

# Show marked image
cv2.imshow("Marked Image", marked)
cv2.waitKey(0)
cv2.destroyAllWindows()

print(center_x,center_y)
```

```python
x = center_x
y = center_y

def Accident_location (x, y):
    if x > 54 and x < 78 and y > 30 and y < 140:
        return "left side car"
    elif x > 105 and x < 130 and y > 30 and y < 140:
        return "right side car"
    elif x > 40 and x < 130 and y > -100 and y < 34:
        return "Front of car"
    elif x > 60 and x < 135 and y > 140 and y < 200:
        return "Back of car"
    elif x > 137 and x < 163 and y > 46 and y < 145:
        return "Left side van"
    elif x > 209 and x < 235 and y > 46 and y < 145:
        return "right side van"
    elif x > 140 and x < 235 and y > -100 and y < 35:
        return "Front of van"
    elif x > 135 and x < 235 and y > 140 and y < 200:
        return "Back of van"
    elif x > -10 and x < 20 and y > 73 and y < 130:
        return "left side bike"
    elif x > 21 and x < 42 and y > 73 and y < 130:
        return "right side bike "
    elif x > 1 and x < 40 and y > 0 and y < 70:
        return "front of bike"
    elif x > 1 and x < 40 and y > 135 and y < 200:
        return "Back of bike"
    else:
        return "ERROR"


print(Accident_location(x, y), "A")  
```
